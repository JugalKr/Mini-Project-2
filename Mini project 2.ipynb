#### exponential fit curve for cases 

```python
def exp_fit(x,a,b,c):
    y = a * np.exp(b*x)+c
    return y
```

```python
fit = curve_fit(exp_fit, df2["days"][:200],df2["tot_cases"][:200],p0 = [0.005, 0.03, 5])
fit_eq = fit[0][0]*np.exp(fit[0][1]*df2["days"][:200])+ fit[0][2]
```

```python
fig =plt.figure()
ax = fig.subplots()
ax.scatter(df2["days"][:200],df2["tot_cases"][:200],color='b', s = 30)
ax.plot(df2["days"][:200], fit_eq, color='r', alpha =1)
ax.set_ylabel('Total Cases')
ax.set_xlabel('No of Days')
plt.show()
```
